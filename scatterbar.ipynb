```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
df=pd.read_stata("/content/sample_data/pm.dta")
df.head(2)
```

```python
df.isnull().sum()
```

```python
x_column='pm1'
y_column='temp'
```

```python
plt.scatter(df[x_column],df[y_column])
plt.title("Correlation between pollution vs temperature")
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.show()
```

```python
df.corr()
```

```python
df["Nor_temp"] = df["temp"]  / df["temp"].abs().max()
df["Nor_pm1"] = df["pm1"]  / df["pm1"].abs().max()
```

```python
plt.scatter(df['Nor_pm1'], df['Nor_temp'])
plt.title("Correlation between pollution vs temperature")
plt.ylabel("Nor_temp")
plt.xlabel("Nor_pm1")
plt.show()
```

```python
x_month='mo'
y_pollution='pm1'
```

```python
plt.scatter(df[x_month],df[y_pollution])
plt.title("Correlation between pollution vs month")
plt.xlabel(x_month)
plt.ylabel(y_pollution)
plt.show()
```

```python

x_gastro='gasyng'
y_pollution='pm1'
plt.scatter(df[x_gastro],df[y_pollution])
plt.title("Correlation between Pollution vs gastro intestinal disease")
plt.xlabel(x_gastro)
plt.ylabel(y_pollution)
plt.show()
```
